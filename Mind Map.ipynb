```python
import matplotlib.pyplot as plt
import networkx as nx

# Créer un graphe orienté
G = nx.DiGraph()

# Ajouter les nœuds
G.add_node("Gestion de l'Énergie")
G.add_node("Consommation énergétique")
G.add_node("Optimisation énergétique")
G.add_node("Modélisation et Prédiction")
G.add_node("Facteurs influents")
G.add_node("Mesures de consommation")
G.add_node("Techniques d'optimisation")
G.add_node("Énergie renouvelable")
G.add_node("Régression linéaire")
G.add_node("Machine Learning")

# Ajouter les relations
G.add_edges_from([
    ("Gestion de l'Énergie", "Consommation énergétique"),
    ("Gestion de l'Énergie", "Optimisation énergétique"),
    ("Gestion de l'Énergie", "Modélisation et Prédiction"),
    ("Consommation énergétique", "Facteurs influents"),
    ("Consommation énergétique", "Mesures de consommation"),
    ("Optimisation énergétique", "Techniques d'optimisation"),
    ("Optimisation énergétique", "Énergie renouvelable"),
    ("Modélisation et Prédiction", "Régression linéaire"),
    ("Modélisation et Prédiction", "Machine Learning")
])

# Dessiner le graphe
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)  # Positionnement des nœuds
nx.draw_networkx(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold')
plt.title("Carte Mentale - Gestion de l'Énergie", fontsize=15)
plt.show()
```
