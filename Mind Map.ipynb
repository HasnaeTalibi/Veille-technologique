```python
import matplotlib.pyplot as plt
import networkx as nx

# Crer un graphe orient
G = nx.DiGraph()

# Ajouter les nuds
G.add_node("Gestion de l'nergie")
G.add_node("Consommation nergtique")
G.add_node("Optimisation nergtique")
G.add_node("Modlisation et Prdiction")
G.add_node("Facteurs influents")
G.add_node("Mesures de consommation")
G.add_node("Techniques d'optimisation")
G.add_node("nergie renouvelable")
G.add_node("Rgression linaire")
G.add_node("Machine Learning")

# Ajouter les relations
G.add_edges_from([
    ("Gestion de l'nergie", "Consommation nergtique"),
    ("Gestion de l'nergie", "Optimisation nergtique"),
    ("Gestion de l'nergie", "Modlisation et Prdiction"),
    ("Consommation nergtique", "Facteurs influents"),
    ("Consommation nergtique", "Mesures de consommation"),
    ("Optimisation nergtique", "Techniques d'optimisation"),
    ("Optimisation nergtique", "nergie renouvelable"),
    ("Modlisation et Prdiction", "Rgression linaire"),
    ("Modlisation et Prdiction", "Machine Learning")
])

# Dessiner le graphe
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)  # Positionnement des nuds
nx.draw_networkx(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold')
plt.title("Carte Mentale - Gestion de l'nergie", fontsize=15)
plt.show()
```
